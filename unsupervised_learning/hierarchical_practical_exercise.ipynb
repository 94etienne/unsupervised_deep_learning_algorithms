```python
# ======================
# 1. IMPORT LIBRARIES
# ======================
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
# ======================
# 2. DATA GENERATION
# ======================
# Generate synthetic data (200 points, 4 centers)
dataset = make_blobs(n_samples=200,
                     n_features=2,
                     centers=4,
                     cluster_std=1.6,
                     random_state=50)

points = dataset[0]  # Only X values
```

```python
# ======================
# 3. DENDROGRAM ANALYSIS
# ======================
# Plot dendrogram to visualize cluster structure
plt.figure(figsize=(10, 5))
dendrogram = sch.dendrogram(sch.linkage(points, method='ward'))
plt.title("Dendrogram")
plt.xlabel("Samples")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
# ======================
# 4. CLUSTERING MODEL
# ======================
# Perform Agglomerative Clustering
hc = AgglomerativeClustering(n_clusters=4, metric='euclidean', linkage='ward')
y_hc = hc.fit_predict(points)
```

```python
# ======================
# 5. VISUALIZATION
# ======================
# Plot clusters
plt.figure(figsize=(8, 6))
plt.scatter(points[y_hc == 0, 0], points[y_hc == 0, 1], s=100, c='cyan', label='Cluster 1')
plt.scatter(points[y_hc == 1, 0], points[y_hc == 1, 1], s=100, c='yellow', label='Cluster 2')
plt.scatter(points[y_hc == 2, 0], points[y_hc == 2, 1], s=100, c='red', label='Cluster 3')
plt.scatter(points[y_hc == 3, 0], points[y_hc == 3, 1], s=100, c='green', label='Cluster 4')
# ======================
# 6. CENTROID CALCULATION
# ======================
# Compute and plot centroids manually (Option)
for i in range(4):
    cluster_points = points[y_hc == i]
    centroid = cluster_points.mean(axis=0)
    print(f"Centroid of Cluster {i + 1}: {centroid}")
    plt.scatter(centroid[0], centroid[1], s=250, c='black', marker='X', edgecolor='white', linewidth=1.5)
    # ======================
# 7. FINAL PLOT SETTINGS
# ======================
plt.title("Agglomerative Clustering with Centroids")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```
